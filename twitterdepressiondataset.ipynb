```python
def cleandata(df):
    df_numeric = df.select_dtypes(include=[np.number])
    Q1 = df_numeric.quantile(0.25)
    Q3 = df_numeric.quantile(0.75)
    IQR = Q3 - Q1
    mask = ~((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR))).any(axis=1)
    return df.loc[mask]
```

```python
import matplotlib.pyplot as plt
def plot_model_comparison(datasets, dataset_names, model_names):
    metrics = ['Mean Absolute Error', 'Mean Squared Error', 'Root Mean Squared Error']
    for metric in metrics:
        plt.figure(figsize=(10, 6))
        for i, dataset_name in enumerate(dataset_names):
            values = [0.5, 0.2, 0.3]
            plt.bar([f'{model} ({dataset_name})' for model in model_names], values, label=dataset_name)
        plt.xlabel("Models")
        plt.ylabel(metric)
        plt.title(f"Comparison of {metric} across Datasets and Models")
        plt.xticks(rotation=45, ha='right')
        plt.legend()
        plt.tight_layout()
        plt.show()
plot_model_comparison([df2, df_cleaned, df_clean], ["df2", "df_cleaned", "df_clean"], ["SVM", "Random Forest", "XGBoost"])
```

```python
def skewness(data):
    n = len(data)
    if n < 3:
        raise ValueError("Dataset must have at least 3 data points for skewness calculation.")
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)
    if std_dev == 0:
        return 0
    sum_of_third_powers = np.sum([(x - mean)**3 for x in data])
    skewness_value = (n / ((n - 1) * (n - 2))) * (sum_of_third_powers / (std_dev**3))
    return skewness_value
for df_name, df in zip(['df_cleaned', 'df2', 'df_clean'], [df_cleaned, df2, df_clean]):
    numerical_cols = df.select_dtypes(include=np.number).columns
    print(f"Skewness for {df_name}:")
    for col in numerical_cols:
        try:
            skew = skewness(df[col].values)
            print(f"{col}: {skew}")
        except ValueError as e:
            print(f"Error calculating skewness for {col} in {df_name}: {e}")
    print("\n")
```

```python
def kurtosis(data):
    n = len(data)
    if n < 4:
      raise ValueError("Dataset must have at least 4 data points for kurtosis calculation.")
    mean = np.mean(data)
    std_dev = np.std(data)
    if std_dev == 0:
        return 0
    sum_of_fourth_powers = np.sum([(x - mean)**4 for x in data])
    kurtosis_value = (n * (n + 1) / ((n - 1) * (n - 2) * (n - 3))) * (sum_of_fourth_powers / (std_dev**4)) - (3 * (n - 1)**2) / ((n - 2) * (n - 3))
    return kurtosis_value
for df_name, df in zip(['df_cleaned', 'df2', 'df_clean'], [df_cleaned, df2, df_clean]):
    numerical_cols = df.select_dtypes(include=np.number).columns
    for col in numerical_cols:
        try:
            kurt = kurtosis(df[col].values)
            print(f"Kurtosis for {col} in {df_name}: {kurt}")
        except ValueError as e:
            print(f"Error calculating kurtosis for {col} in {df_name}: {e}")
```
